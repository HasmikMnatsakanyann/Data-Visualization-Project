### Visualizations

Unique values for some columns

```python
values = df["Continent"].value_counts().values
keys = df["Continent"].value_counts().keys()

colors = ['#66c2a5','#fc8d62','#8da0cb','#e78ac3','#a6d854','#ffd92f']

cont_num = pd.DataFrame({"Countinents": keys, "Quantity": values})
sns.barplot(data=cont_num, x="Quantity", y="Countinents", palette = colors)
```

```python
values = df["DevName"].value_counts().values
keys = df["DevName"].value_counts().keys()
dn = pd.DataFrame({"DevName": keys, "Quantity": values})
color = ['#fc8d62','#8da0cb']
sns.barplot(data=dn, x="DevName", y="Quantity", palette = color)
```

Encode "DevName" values to 0 and 1

```python
temp = df["DevName"].copy()
temp[df["DevName"]=="Developing regions"]=0
temp[df["DevName"]=="Developed regions"]=1
df["DevName"] = temp
df
```

The proportion of zeros

```python
temp = {}
for i in df.columns[4:38]:
  temp[i] = df[i].value_counts()[0]
keys = temp.keys()
values = temp.values()
data = pd.DataFrame({"Year": keys, "NumOfZeroes": values})

fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(1, 1, 1)

ax.barh(y=data["Year"], width=data["NumOfZeroes"], color='#8da0cb')

plt.show()
```

The distribution of "Total" with respect to "Country"

```python
dens= pd.DataFrame(df.groupby("Country")["Total"].sum())

sns.kdeplot(data=dens, x="Total", color = '#377eb8')
```

The distribution of "Total" with respect to "Region"

```python
denst = pd.DataFrame(df.groupby("Region")["Total"].sum())

sns.displot(denst["Total"], kde=False, color = '#fc8d62')
```

Correlation heatmap for 2000's

```python
columns1 = df.columns[24:34]

dataplot = sns.heatmap(df[columns1].corr(), cmap="magma", annot=True)
```

```python
import plotly.express as px
import plotly.graph_objects as go

sum_years = df[df.columns[4:38]].sum()
values = sum_years.values
keys = sum_years.keys()
sum_years = pd.DataFrame({"Years": keys, "Quantity": values})

# create a line plot
fig = go.Figure()
fig.add_trace(go.Scatter(x=sum_years['Years'], y=sum_years['Quantity'], mode='lines'))

# set the title and axis labels
fig.update_layout(title='Number of Immigrants from 1980-2013', xaxis_title='Year', yaxis_title='Number Of Immigrants')

# show the plot
fig.show()
```
